```python
# make a plot of the probability density function
import numpy as np
import matplotlib.pyplot as plt

xpoint = 3
ypoint = 0.5
print(DM.get_prob_3d(xpoint = 0, ypoint = 0.5, time = 34, convolve = False)[0])

time = np.arange(0, 200, 1)
prob = np.zeros(200)
for i in range(200):
    prob[i] = DM.get_prob_3d(xpoint = xpoint, ypoint = ypoint, time = i, convolve = True)[0]

plt.plot(time, prob)
plt.show()
```
